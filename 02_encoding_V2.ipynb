```python
import pandas as pd
from music21 import converter, instrument, note, chord
from sklearn.preprocessing import LabelEncoder
import itertools
```

```python
path = 'data\maestro-v3.0.0.csv'
df = pd.read_csv(path)
df = df.sort_values(by='duration').iloc[:2]
```

```python
df['midi_filename'].iloc[0]
```

```python
from music21 import converter

score = converter.parse("data/"+df['midi_filename'].iloc[0])
L_notes = []
for element in score.recurse().notes:
    if element.isNote:
        L_notes.append([element.pitch.midi])
    elif element.isChord:
        L_notes.append([p.midi for p in element.pitches])
```

```python
L_notes2 = []
for k in L_notes:
    while len(k)<10:
        k.append(-1)
    L_notes2.append(k)
```

```python
import pandas as pd
from music21 import chord
from itertools import product
```

```python
all_notes = []
doigts=1
```

```python
n = 10  # nombre de niveaux
base = range(5)

mapping = {}
all_notes = []
i=0
for k in range(1,n+1):
    all_notes += [list(t) for t in product(base, repeat=k)]

for k in all_notes:
    mapping[chord.Chord(k)]=i
    i+=1
```
